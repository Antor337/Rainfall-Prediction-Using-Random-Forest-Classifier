```python
import pandas as pd
```

```python
df=pd.read_csv('dataset/rainfall.csv')
```

```python
df
```

```python
def newvalue(Rainfall):
    if Rainfall==0.0:
        return 0
    if Rainfall>0.0 and Rainfall<=2.0:
        return 1
    if Rainfall>2.0 and Rainfall<11.0:
        return 2
    if Rainfall >=11.0:
        return 3
```

```python
df['Rainfall']=df['Rainfall'].apply(newvalue)
df
```

```python
df.to_csv('dataset/rainfall.csv')
```
